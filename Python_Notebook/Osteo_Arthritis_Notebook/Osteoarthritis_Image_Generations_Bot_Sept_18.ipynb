```python
# Input prompt
# Define the main prompt
output_list = []
# main_prompt = (
#     "Generate a diverse set of 50 text prompts to create only X-ray images of knee depicting individuals of varying ages afflicted by knee Osteoarthritis. "
#     "Ensure that the images vividly portray the knee joint with evident indications of Osteoarthritis arthritis. "
#     "The primary aim is to generate high-quality X-ray images that accurately represent the condition, thereby aiding those who are suffering. "
#     "Capture different degrees of severity and variations in joint inflammation, erosion, and deformity to provide comprehensive visual assistance."
# )

# main_prompt = (
# "Generate 100 textual prompts to generate X-ray images depicting individuals of various ages, from young adults to seniors, each showcasing different stages and manifestations of osteoarthritis in their knee joints." 
# "Ensure that the X-ray images are of high quality and accurately represent the condition. "
# "Provide a diverse range of scenarios and age groups to cover the spectrum of osteoarthritis cases"
# )

main_prompt = (
"Generate 100 unique textual prompts to produce high-quality X-ray images specifically focusing on either the left leg knee or right knee of individuals across different age groups, ranging from young adults to seniors." 
"Each prompt should vividly illustrate various stages and manifestations of osteoarthritis in the specified knee joint."
"Please ensure that the images accurately represent the condition, covering a diverse range of scenarios and age groups to encompass a comprehensive spectrum of osteoarthritis cases."
)

# Generate responses using the OpenAI API
response = openai.Completion.create(
    engine="text-davinci-003",
    prompt=main_prompt,
    max_tokens=3000,
    n=1,
    stop=None,
    temperature=0.2
)

#print(response)

# Extract the generated output
output = response.choices[0].text.strip()
print(output)

output_list.append(output.strip())






# # You can now use the 'output_list' for further processing
```

```python

# Split the text by newline characters and remove empty strings
text_lines = [line.strip() for line in output_list[0].split('\n') if line.strip()]
# text_lines = text_lines[1:]
# text_lines

# Remove numbers and full stops
cleaned_prompts_osteo = [line.split('. ',1)[1] for line in text_lines]

cleaned_prompts_osteo
```

```python
def download(url, file_name):
    # open in binary mode
    with open(file_name, "wb") as file:
        # get request
        response = get(url)
        # write to file
        file.write(response.content)
```
